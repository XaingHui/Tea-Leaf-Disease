```python
import torch
import torchvision
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split

# 1. 数据加载
data_dir = './tea sickness dataset'  # 数据集目录
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # 调整图像大小
    transforms.ToTensor(),  # 转换为张量
])

print("Loading dataset...")
dataset = ImageFolder(root=data_dir, transform=transform)

# 输出一些基本信息
print(f"Number of images: {len(dataset)}")
print(f"Classes: {dataset.classes}")

# 2. 数据标签
# ImageFolder会自动为每个子目录分配整数标签

# 3. 数据分割为训练集和测试集
train_size = int(0.7 * len(dataset))
test_size = len(dataset) - train_size

train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

# 输出数据集分割信息
print(f"Training dataset size: {len(train_dataset)}")
print(f"Testing dataset size: {len(test_dataset)}")

# 创建数据加载器
batch_size = 32

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size)

# 输出数据加载器信息
print(f"Number of batches in training dataloader: {len(train_loader)}")
print(f"Number of batches in testing dataloader: {len(test_loader)}")
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
import matplotlib.pyplot as plt
import random
import os
# 加载预训练的MobileNetV2模型
mobilenet = models.mobilenet_v2(pretrained=True)

# 冻结所有层的参数，以便微调
for param in mobilenet.parameters():
    param.requires_grad = False

# 获取模型的分类层输入特征数
num_ftrs = mobilenet.classifier[1].in_features

# 替换模型的分类层以适应您的类别数
num_classes = len(dataset.classes)  # 根据您的数据集类别数修改
mobilenet.classifier[1] = nn.Linear(num_ftrs, num_classes)

# 如果需要，将模型转移到GPU
device = torch.device("cpu")
mobilenet = mobilenet.to(device)

# 定义损失函数和优化器
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(mobilenet.parameters(), lr=0.001, momentum=0.9)
# 初始化一个空列表，用于记录每个训练周期的损失值
train_losses = []

# 训练模型
num_epochs = 10  # 根据需要修改
for epoch in range(num_epochs):
    mobilenet.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = mobilenet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}")
    train_losses.append(running_loss/len(train_loader))
    # 保存模型在每个训练周期结束后
    model_save_path = f"mobilenet_epoch_{epoch+1}.pt"
    torch.save(mobilenet.state_dict(), model_save_path)
    print(f"Model saved at {model_save_path}")
    # 保存模型权重
model_save_path = 'mobilenet_weights.pth'
torch.save(mobilenet.state_dict(), model_save_path)
    #mobilenet.load_state_dict(torch.load('MobileNetModel.pth'))

# 绘制损失函数图
plt.plot(range(1, num_epochs + 1), train_losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.legend()
plt.show()
# 在测试集上评估模型
mobilenet.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = mobilenet(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Accuracy on the test set: {100 * correct / total}%")


# 设置随机种子以获取可重复的结果
random.seed(0)

# 数据加载
data_dir = './tea sickness dataset'  # 数据集目录
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # 调整图像大小
    transforms.ToTensor(),  # 转换为张量
])

# 创建一个包含所有类别目录的列表
class_directories = [os.path.join(data_dir, class_name) for class_name in os.listdir(data_dir)]

# 初始化一个空列表，用于存储随机抽取的图像路径
random_images = []

# 从每个类别中随机抽取两张图片
for class_dir in class_directories:
    image_files = os.listdir(class_dir)
    if len(image_files) >= 2:
        random_images.extend(random.sample(image_files, 2))

# 定义类别名称映射
class_names = dataset.classes

# 循环遍历随机抽取的图像并进行识别和展示
for image_file in random_images:
    image_path = os.path.join(data_dir, image_file)
    image = Image.open(image_path)
    image = transform(image)
    image = image.unsqueeze(0)  # 添加批次维度

    # 推理
    with torch.no_grad():
        outputs = mobilenet(image)
        predicted = torch.max(outputs, 1)

    # 获取类别名
    class_index = predicted.item()
    class_name = class_names[class_index]

    # 展示图像和识别结果
    plt.imshow(image.squeeze(0).permute(1, 2, 0))
    plt.title(f"Predicted Class: {class_name}")
    plt.axis('off')
    plt.show()
```
